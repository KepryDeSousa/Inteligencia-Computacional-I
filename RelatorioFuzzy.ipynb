```python
import numpy as np 
import pandas as pd
!pip install scikit-fuzzy
import skfuzzy as fuzz
from skfuzzy import control as crtl
```

### O Que é logica Fuzzy 
> * Um Sistema de Inferência Fuzzy (SIF) é um tipo especial de Sistema Baseado em Conhecimento (SBC). A Base de Conhecimento é composta de um conjunto de regras fuzzy. A Máquina de Inferência implementa um mecanismo de inferência fuzzy.

Referencia

```python
#Antecedentes 

#Velocidade
Distancia = crtl.Antecedent(np.arange(0,100,0.5),"Distancia")

##################################################
Distancia['LE'] = fuzz.trapmf(Distancia.universe,[0,0,10,35])
Distancia['LC'] = fuzz.trimf(   Distancia.universe,[30,40,50])
Distancia['CE'] = fuzz.trimf(   Distancia.universe,[45,50,55])
Distancia['RC'] = fuzz.trimf(   Distancia.universe,[50,60,70])
Distancia['HI']  = fuzz.trapmf(Distancia.universe,[65,90,100,100])
Distancia.universe
```

```python
Angulo    =  crtl.Antecedent(np.arange(0,360,0.5),"Angulo")
Angulo.universe.view()
```

```python
#Angulo

Angulo['RB'] = fuzz.trimf(Angulo.universe,[ 0, 55,100])
Angulo['RU'] = fuzz.trimf(Angulo.universe,[90,125,145])
Angulo['RV'] = fuzz.trimf(Angulo.universe,[135,165,180])
Angulo['VE'] = fuzz.trimf(Angulo.universe,[170,180,190])
Angulo['LV']  = fuzz.trimf(Angulo.universe,[180,200,225])
Angulo['LU'] = fuzz.trimf(Angulo.universe,[215,235,270])
Angulo['LB'] = fuzz.trimf(Angulo.universe,[260,290,360])




```

```python
#Consequente da logica Fuzzy

delta = crtl.Consequent(np.arange(0,60,1),"Delta")
delta.universe
```

#Função Triangular 

```python
Angulo.view(),
Distancia.view()
```

```python
#nb , nm , ns , ze , ps,  pm,  pb
delta['NB'] = fuzz.trimf(delta.universe,[ 0,0,15])
delta['NM'] = fuzz.trimf(delta.universe,[5,15,25])
delta['NS'] = fuzz.trimf(delta.universe,[20,25,30])
delta['ZE'] = fuzz.trimf(delta.universe,[25,30,35])
delta['PS']  = fuzz.trimf(delta.universe,[30,35,40])
delta['PM'] = fuzz.trimf(delta.universe,[35,45,55])
delta['PB'] = fuzz.trimf(delta.universe,[ 45,55,60])

delta.view()
```

```python
#Desenvolvimento das regras
Regra_1 = crtl.Rule( (Distancia["LE"]  & Angulo["RB"]) | (Distancia['LE'] & Angulo["RU"]), delta['NB' ])

Regra_2 = crtl.Rule( (Distancia["LC"] & Angulo["RU"])  | (Distancia['LE'] &  Angulo["RU"]) | (Distancia["RC"] & Angulo["RV"]), delta['NM']  )

Regra_3 = crtl.Rule(  (Distancia["LC"] & Angulo["RU"])  | (Distancia["CE"] & Angulo["VE"] )  | (Distancia["RC"] & Angulo["RV"]), delta['NS'])

Regra_4 = crtl.Rule(  (Distancia["RC"] & Angulo["RV"]) | (Distancia["LC"] &  Angulo["LV"]) | (Distancia["CE"] & Angulo["VE"] ) , delta['ZE'])

Regra_5  =crtl.Rule(  (Distancia["CE"] & Angulo["VE"] )   | (Distancia["CE"] & Angulo["VE"] )  |  (Distancia["RC"] &  Angulo["LU"]) , delta["PS"] )

Regra_6 =crtl.Rule( (Distancia["HI"]  & Angulo["LB"] ) |  (Distancia["RC"] &  Angulo["LU"]) |  (Distancia["RC"] &  Angulo["LV"]) , delta["PM"] )

Regra_7 =crtl.Rule( (Distancia["HI"]  &  Angulo["LB"]) ,  delta['PB'])
```

```python
#Desenvolvendo o sistema com as regras ja associadas
System_Control = crtl.ControlSystem([Regra_1,Regra_2,Regra_3,Regra_4,Regra_5,Regra_6,Regra_7]) 
System= crtl.ControlSystemSimulation(System_Control)
```

```python
#Chamando os valores 
System.input['Distancia']   = 47.5
System.input['Angulo']  =99.5
System.compute() 
```

```python
#Desfuzificação 
print(System.output)
```

```python
delta.view(sim = System),
Angulo.view(sim = System),
Distancia.view(sim = System)
```

# Função Gaussiana 

```python
#Antecedentes 

#Distancia
Distancia = crtl.Antecedent(np.arange(0,100,0.5),"Distancia")

##################################################
Distancia['LE'] = fuzz.gaussmf(Distancia.universe,10,35)
Distancia['LC'] = fuzz.gaussmf(   Distancia.universe,40,50)
Distancia['CE'] = fuzz.gaussmf(   Distancia.universe,45,55)
Distancia['RC'] = fuzz.gaussmf(   Distancia.universe,50,70)
Distancia['HI']  = fuzz.gaussmf(Distancia.universe,65,100)
Distancia.view()
```

```python
Angulo    =  crtl.Antecedent(np.arange(0,360,0.5),"Angulo")
Angulo.universe.view()
```

```python
#Angulo

Angulo['RB'] = fuzz.gaussmf(Angulo.universe, 0,100)
Angulo['RU'] = fuzz.gaussmf(Angulo.universe,90,145)
Angulo['RV'] = fuzz.gaussmf(Angulo.universe,135,180)
Angulo['VE'] = fuzz.gaussmf(Angulo.universe,170,190)
Angulo['LV']  = fuzz.gaussmf(Angulo.universe,180,225)
Angulo['LU'] = fuzz.gaussmf(Angulo.universe,215,270)
Angulo['LB'] = fuzz.gaussmf(Angulo.universe,260,360)
Angulo.view()
```

```python
#Consequente da logica Fuzzy

delta = crtl.Consequent(np.arange(0,60,1),"Delta")
delta.universe

delta['NB'] = fuzz.gaussmf(delta.universe, 0,15)
delta['NM'] = fuzz.gaussmf(delta.universe,5,25)
delta['NS'] = fuzz.gaussmf(delta.universe,20,30)
delta['ZE'] = fuzz.gaussmf(delta.universe,25,35)
delta['PS']  = fuzz.gaussmf(delta.universe,30,40)
delta['PM'] = fuzz.gaussmf(delta.universe,35,55)
delta['PB'] = fuzz.gaussmf(delta.universe, 45,60)
```

```python
#Desenvolvimento das regras
Regra_1 = crtl.Rule( (Distancia["LE"]  & Angulo["RB"]) | (Distancia['LE'] & Angulo["RU"]), delta['NB' ])

Regra_2 = crtl.Rule( (Distancia["LC"] & Angulo["RU"])  | (Distancia['LE'] &  Angulo["RU"]) | (Distancia["RC"] & Angulo["RV"]), delta['NM']  )

Regra_3 = crtl.Rule(  (Distancia["LC"] & Angulo["RU"])  | (Distancia["CE"] & Angulo["VE"] )  | (Distancia["RC"] & Angulo["RV"]), delta['NS'])

Regra_4 = crtl.Rule(  (Distancia["RC"] & Angulo["RV"]) | (Distancia["LC"] &  Angulo["LV"]) | (Distancia["CE"] & Angulo["VE"] ) , delta['ZE'])

Regra_5  =crtl.Rule(  (Distancia["CE"] & Angulo["VE"] )   | (Distancia["CE"] & Angulo["VE"] )  |  (Distancia["RC"] &  Angulo["LU"]) , delta["PS"] )

Regra_6 =crtl.Rule( (Distancia["HI"]  & Angulo["LB"] ) |  (Distancia["RC"] &  Angulo["LU"]) |  (Distancia["RC"] &  Angulo["LV"]) , delta["PM"] )

Regra_7 =crtl.Rule( (Distancia["HI"]  &  Angulo["LB"]) ,  delta['PB'])
```

```python
#Chamando o sistema de controle
System_Control = crtl.ControlSystem([Regra_1,Regra_2,Regra_3,Regra_4,Regra_5,Regra_6,Regra_7]) 
System= crtl.ControlSystemSimulation(System_Control)
```

### Resultados 
note que os resultados da Fuzzyficação estao logo na primeira linha, anteriormente aos graficos


```python
# Passando os dados do sistema 
System.input['Distancia']   = 47.5
System.input['Angulo'] =  189
System.compute() 

#Graficos de Resposta
Distancia.view(sim = System)
Angulo.view(sim = System)
delta.view(sim = System)
#Defuzzyficação 
#note que esse valor sai antes dos graficos
print(f"Linha 1: {System.output}")
```
