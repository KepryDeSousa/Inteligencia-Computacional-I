### População Inicial 
> Inicialização: criação de uma população inicial de soluções candidatas de forma aleatória.

```python

class Populacao:
    def __init__(self, n_pop, n_cidades):
        self.n_pop = n_pop
        self.n_Geracoes = n_pop
        self.n_cidades = n_cidades
        

    def Cidades_Visitadas(self):
        cidades = list(range(self.n_cidades))
        random.shuffle(cidades)
        cidades.append(cidades[0])
        return cidades
    
    def Cromossomo(self):
        pop = [self.Cidades_Visitadas() for i in range(self.n_pop)]
        return pop
    
```

### Avaliação

```python
import math 
import statistics
```

```python
# função de distância entre duas cidades
def Distancia(city1, city2):
    return math.sqrt((city1[1] - city2[1]) ** 2 + (city1[2] - city2[2]) ** 2)

def Custo_Populacao(dado):
    d=[]
    for i in range(1,len(dado),1): 
      d.append(Distancia(bd.iloc[dado[i-1]],bd.iloc[dado[i]]))
   
    d = statistics.mean(d)
    return d    
# Esta função pega o cromossomo enviado e devolve a media de todas as distancias
# calulada em Distancia(Funcao Euclidiana)


def Tabela_Custo(pop):
  a = []
  for i in range(len(pop)):
    a.append(Custo_Populacao(pop[i]))
  custo = pd.DataFrame(a,columns=["Custo"])
  
  return custo
# Esta função pega o cromossomo enviado e devolve a media de todas as distancias
# calulada em Distancia(Funcao Euclidiana)


# houve erro ao tentar trazer esse trecho para o OO para terminar o trabalho a tempo
# voltei o cod para o mais simples
```

### Seleção

Os codigos de crossover não chegaram a um fim. o problema vinha do momento de 
acertar os valores sem repetições dentro da lista sem perder a identidade
muitos metodos foram usados e muitos deles devolviam a lista organizada(sem identidade propria )
no livro "Introduction To Evolutionary Computing" do J.E. Smith me deu a ideia de trazer um algoritmo 
Assexuado sem o cruzamento dos dados 


```python
class Selecao:
  def __init__(self,n_Geracoes,pop,n_cidades):
    self.n_Geracoes = n_Geracoes
    self.lista = []
    self.pop = pop
    self.n_cidades = n_cidades

  def Selecionar(self):
     lista = []
     for i in range(self.n_Geracoes):
       pop = Populacao(self.pop,self.n_cidades)
       pop = pop.Cromossomo()
       a = pd.DataFrame(pop)
       # Ate aqui definimos a nova pop

       tabela = Tabela_Custo(pop)
       a['Custo'] = tabela
       a['Geracao'] = i
       lista.append(a)
     return pd.concat(lista, ignore_index=True)
```

```python
class Eletismo:
   def __init__(self,dados,n_Geracoes):
    self.dados = dados
    self.n_Geracoes = n_Geracoes
   
   def N_Elite(self):
     lista_Napt = self.dados.sort_values('Custo')
     lista_Napt = lista_Napt[900:1000]
     return lista_Napt

   def Elite(self):
     lista_apt = self.dados.sort_values('Custo')
     lista_apt = lista_apt[0:100]
     return lista_apt

#Selecionando apenas o 100 mais aptos 
# das ultimas 10 G
```

### Mutação
No Algoritmo Assexuado um "CrossOver" possivel é a mutação

```python

class Mutacao:
  def __init__(self, dados):
    self.dados = dados.copy()
    self.mutados = [ ]
  def M(self):
    for i in range(len(self.dados)):
      if random.random() > 0.01:  # alterei a condição para permitir mais mutações
          self.mutados = self.dados.drop(i, inplace=True)
          #fazendo uma lista com os indexs das cidades randomicas
          
    return self.mutados
```

```python
M = Mutacao(apt)
Mutados = M.M()
print(Mutados)
```
