### Custo 


```python
# função de distância entre duas cidades
def Distancia(city1, city2):
    return math.sqrt((city1[0] - city2[0]) ** 2 + (city1[1] - city2[1]) ** 2)
```

### Particula

```python
class Particula():
  def __init__(self,range1, range2):
    self.range1 = range1
    self.range2 = range2
    
  def Rotas(self):
    rota1 = [random.randint(self.range1,self.range2), random.randint(self.range1,self.range2)]
    rota2 = [random.randint(self.range1,self.range2), random.randint(self.range1,self.range2)]
    rota3 = [random.randint(self.range1,self.range2), random.randint(self.range1,self.range2)]
    # construindo os valores das particulas 

    Rota_Custo_I = Distancia(rota1,rota2)
    Rota_Custo_II = Distancia(rota2,rota3)


    velocidade = 0
    data = [rota1,rota2,rota3,Rota_Custo_I,Rota_Custo_II,velocidade]
    return data
```

```python
p = Particula(0,10)
p.Rotas()
```

### Bando de Aves



```python
class Bando():
  def __init__(self, particulas, range1,range2):
    self.range1 = range1
    self.range2 = range2
    self.particulas = particulas

  def Revoada(self):
    lista = []
    for i in range(self.particulas):
      p = Particula(self.range1,self.range2)
      #sempre inicio uma nova particula
      lista.append(p.Rotas())
      data = pd.DataFrame(lista)
      data = data.rename(columns={0: 'Caminho_I', 1: 'Caminho_II',2: 'Caminho_III',3:"Distancia_I",4:"Distancia_II",5:"Velocidade"})
    
    return data
```

```python
b = Bando(100,0,100)
c = b.Revoada()
c
```

### Local_best x Global_best

```python
class Locais():
  def __init__(self, banco):
    self.banco = banco

  def Locais_L(self):
    Local_Best = []
    for i in range(self.banco.shape[0]):
      if self.banco.loc[i].Distancia_I < self.banco.loc[i].Distancia_II:
        Local_Best.append(self.banco.loc[i].Distancia_I)
      else:
        Local_Best.append(self.banco.loc[i].Distancia_II)
     
    self.banco['Local_Best'] = Local_Best
    return self.banco
  
  def Local_Best_Global(self):
    Global_Best = []
    self.banco['Global_Best'] = self.banco.Local_Best.min()
    return self.banco
```

```python
Loc = Locais(c)
L = Loc.Locais_L()
L = Loc.Local_Best_Global()
```

### Velocidade

```python
class Velocidade():
  def __init__(self,Bd,r1,r2):
    self.Bd = Bd
    self.r1 = r1
    self.r2 = r2

  def V_Bd(self):
    Velocidade_ = []
    for i in range(self.Bd.shape[0]): 
      vi = self.Bd.loc[i].Velocidade + (2*r1) * (self.Bd.loc[i].Local_Best - i ) + self.Bd.loc[i].Velocidade + (2*r2) * (self.Bd.loc[i].Global_Best - i )
      Velocidade_.append(vi)

    self.Bd["Velocidade"] = Velocidade_
    return self.Bd

class Atualizar_V():
  def __init__(self,Bd):
    self.Bd = Bd.copy()

  def up(self):
    Velocidade_ = []

    vMax = 100.00
    vMin = 0.00001

    for i in range(self.Bd.shape[0]): 
      if vMax < self.Bd.loc[i].Velocidade:
        Velocidade_.append(vMax)

      elif vMin > self.Bd.loc[i].Velocidade:
        Velocidade_.append(vMin)

      else: 
        x = (i + self.Bd.loc[i].Velocidade)
        vi = self.Bd.loc[i].Velocidade + (2*r1) * (self.Bd.loc[i].Local_Best - x ) + self.Bd.loc[i].Velocidade + (2*r2) * (self.Bd.loc[i].Global_Best - x )
        Velocidade_.append(vi)
        continue 

    self.Bd["Velocidade"] = Velocidade_
    return self.Bd

```

## Anckley: Função de Otimização
> Serve para verificar a capacidade um algoritimo ao encontrar o minimo local


```python
from numpy import arange
from numpy import exp
from numpy import sqrt
from numpy import cos
from numpy import e
from numpy import pi


def objective(x, y):
 return -20.0 * exp(-0.2 * sqrt(0.5 * (x**2 + y**2)))-exp(0.5 * (cos(2 * 
  pi * x)+cos(2 * pi * y))) + e + 20

#referencia : https://www.indusmic.com/post/ackley-function
```

### Aplicação de formula


 Primeiro Quartil (Q1):
  >$Q1 = X[((n + 1) / 4)]$

  Terceiro Quartil (Q3):
  > $Q3 = X[((3n + 1) / 4)]$

  Intervalo Interquartil (IQR):
  > $IQR = Q3 - Q1$

  Limite Inferior:
  > $Limite \ Inferior = Q1 - (1.5 * IQR)$

  Limite Superior:
  >$Limite \ Superior = Q3 + (1.5 * IQR)$
